```python
import pandas as pd

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('23_PBP_WHKYHAC_SPORTLOGIQ.csv', encoding='latin-1')
```

```python
df['afterrebound'] = 0
```

```python
filtered_df = df[(df['eventname'] =='shot') & (df['outcome'] == 'successful') & (df['opposing_goalie'].notna())]
```

```python
for _, row in filtered_df.iterrows():
    shot_date = row['date']
    shot_game = row['game']
    end_time = row['compiledgametime']
    start_time = end_time - 5
    if start_time < 0:
        start_time = 0
    last5sec_df = df[(df['date'] == shot_date) &\
                     (df['game'] == shot_game) &\
                     (df['compiledgametime'] < end_time) &\
                     (df['compiledgametime'] >= start_time)]
    if last5sec_df['createsrebound'].any() == True:
        df.loc[row.name, 'afterrebound'] = 1
        filtered_df.loc[row.name, 'afterrebound'] = 1
```

```python
goalie_df = filtered_df[['opposing_goalie', 'xg_all_attempts', 'goal', 'onetimer',
                         'shotaim', 'shottype', 'createsrebound', 'afterrebound',
                         'strengthstate']].reset_index(drop = True)
```

```python
shotaim_gb = goalie_df.groupby('shotaim').sum()
shottype_gb = goalie_df.groupby('shottype').sum()
```

```python
shotaim_gb['aim_goalperxg'] = shotaim_gb['goal']/shotaim_gb['xg_all_attempts']
shottype_gb['type_goalperxg'] = shottype_gb['goal']/shottype_gb['xg_all_attempts']
```

```python
shotaim_list = list(goalie_df['shotaim'].value_counts().index)
```

```python
shottype_list = list(goalie_df['shottype'].value_counts().index)
```

```python
df_length = list(range(0, len(goalie_df)))
```

```python
goalie_df['aim_goalperxg'] = 0
```

```python
goalie_df['type_goalperxg'] = 0
```

```python
for shotaim in shotaim_list:
    for n in df_length:
        if goalie_df.loc[n,]['shotaim'] == shotaim:
            goalie_df['aim_goalperxg'][n] = shotaim_gb.loc[shotaim]['aim_goalperxg']
```

```python
for shottype in shottype_list:
    for n in df_length:
        if goalie_df.loc[n,]['shottype'] == shottype:
            goalie_df['type_goalperxg'][n] = shottype_gb.loc[shottype]['type_goalperxg']
```
